```python
class VowelCounter:
    """Клас для аналізу та обробки голосних літер у тексті."""

    def __init__(self, text: str):
        self.text = text
        self._vowels_set = "aeiouAEIOU"
        self.student_name = "Руслан"
        self.student_group = "КН-32"

    def get_vowels_list(self) -> list:
        
        return [char for char in self.text if char in self._vowels_set]

    def count(self) -> int:
        return len(self.get_vowels_list())

    def process(self) -> str:
        
        found = self.get_vowels_list()
        vowels_str = ", ".join(found) if found else "відсутні"
        return (f"Текст: '{self.text}'\n"
                f"Кількість голосних: {self.count()}\n"
                f"Знайдені літери: {vowels_str}")

    @property
    def vowel_count(self) -> int:
        return self.count()

    def __str__(self) -> str:
        return f"Виконав: {self.student_name}, Група: {self.student_group}"



data_list = ["hello", "world", "Python", "programming", "vowels","aaa"]

for item in data_list:
    counter = VowelCounter(item)
    
    print(counter)           
    print(counter.process()) 
    print("-" * 30)
```

```python
# зробити калькутор для голосних літер, який буде приймати текст і виводити кількість голосних та їх список.

class VowelCounter:
    def __init__(self, text):
        self.text = text
        self.vowels = "aeiouAEIOU"
        self.count = 0
        self.found_vowels = []

    def process(self):
        for char in self.text:
            if char in self.vowels:
                self.count += 1
                self.found_vowels.append(char)
        return self.count, self.found_vowels
# Приклад використання
text = "Hello World!"
counter = VowelCounter(text)
count, vowels_list = counter.process()
print(f"Кількість голосних: {count}")
print(f"Знайдені голосні: {vowels_list}")
```
